```python
import sqlite3
import csv
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine


connection = sqlite3.connect('flight_db.sqlite')
cursor = connection.cursor()
engine = create_engine('sqlite:///flight_db', echo=False)
```

```python
cursor.execute('''
    CREATE TABLE flight_db (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        FL_DATE TEXT,
        DEP_HOUR INTEGER,
        OP_UNIQUE_CARRIER TEXT,
        OP_CARRIER_FL_NUM INTEGER,
        TAIL_NUM TEXT,
        ORIGIN TEXT,
        DEST TEXT,
        SCHED_DEP_DATE DATE,
        SCHED_DEP_TIME TEXT,
        ACTUAL_DEP_DATE DATE,
        ACTUAL_DEP_TIME TEXT,
        TAXI_OUT INTEGER,
        DEP_DELAY INTEGER,
        AIR_TIME INTEGER,
        CANCELLED INTEGER,
        YEAR_OF_MANUFACTURE INTEGER,
        MANUFACTURER TEXT,
        ICAO_TYPE TEXT,
        ACTIVE_WEATHER INTEGER )
''')
```

```python
flight_data = pd.read_csv('FinalAirlineData.csv')             
```

```python
flight_data.to_sql('flight_db', connection, if_exists='append', index=False)
```

```python
cursor.execute('''
    SELECT * FROM flight_db
''')
print(cursor.fetchall())
```

```python
connection.close()
```
